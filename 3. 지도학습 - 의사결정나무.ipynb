```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
# 1. 데이터의 형태 (shape)
df.shape
```

```python
# 2. 데이터의 통계량
df.describe()
```

```python
# 3. 데이터의 정보
df.info()
```

```python
# 결측치 전처리

# 평균값 대체
d_mean = df['Age'].mean()
df['Age'].fillna(d_mean, inplace=True)
```

```python
d_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(d_mode, inplace=True)
```

```python
# 제거

df.drop(columns=['Cabin'], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Embarked'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['en_sex'] = le.fit_transform(df['Sex'])
df['en_embarked'] = le.fit_transform(df['Embarked'])
```

```python
df = pd.concat([df, df['en_sex'], df['en_embarked']], axis=1)
```

```python
df.columns
```

```python
X = df[['Pclass', 'en_sex', 'en_embarked', 'Age']]
y = df['Survived']
```

```python
# 분석 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_predict = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_predict)
print(acc)
```

```python
# 분류를 다양한 모델로 적용해보기

from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_predict = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_predict)
print(acc)
```
