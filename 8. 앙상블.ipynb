```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification, make_blobs
from sklearn.datasets import load_breast_cancer

from matplotlib.colors import ListedColormap
```

```python
fruits = pd.read_table('/content/drive/MyDrive/fruit_data_with_colors.txt')
```

```python
fruits
```

```python
X_fruits = fruits[['height', 'width', 'mass', 'color_score']]
Y_fruits = fruits['fruit_label']
```

```python
Y_fruits.unique()
```

```python
target_names = ['apple', 'mandarin', 'orange', 'lemon']
```

```python
# 랜덤 포레스트로 분류 및 시각화
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_fruits.values, Y_fruits.values, random_state=0)
```

```python
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, Y_train)
```

```python
print(rf_clf.score(X_train, Y_train))
print(rf_clf.score(X_test, Y_test))
```

## 1. bagging을 사용하여 훈련시킨 후 결과를 인쇄하기

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500,
                            max_samples=10, n_jobs=-1, random_state=42)
bag_clf.fit(X_train, Y_train)
```

```python
print(bag_clf.score(X_train, Y_train))
print(bag_clf.score(X_test, Y_test))
```

```python
from sklearn.svm import SVC
svc_clf = BaggingClassifier(estimator=SVC(),n_estimators=10, random_state=0)
svc_clf.fit(X_train, Y_train)
```

```python
print(svc_clf.score(X_train, Y_train))
print(svc_clf.score(X_test, Y_test))
```

## 2. XGboost를 사용하여 훈련시킨 후 결과를 인쇄하기

```python
from xgboost import XGBClassifier
```

```python
Y_train = np.where(Y_train==1, 0, Y_train)
Y_train = np.where(Y_train==2, 1, Y_train)
Y_train = np.where(Y_train==3, 2, Y_train)
Y_train = np.where(Y_train==4, 3, Y_train)
```

```python
model = XGBClassifier()
model.fit(X_train, Y_train)
```

```python
print(model.score(X_train, Y_train))
print(model.score(X_test, Y_test))
```
